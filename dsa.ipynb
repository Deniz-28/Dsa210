```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import pearsonr, ttest_ind, f_oneway
```

```python
df = pd.read_csv("Social Media and Mental Health.csv")
df.head()
```

```python
list(df.columns)
```

```python
df = df.rename(columns={
    'Gender ?': 'Gender',
    'What is your age?': 'Age',
    'On average, how many hours per day do you spend on social media?': 'Usage',
    'How would you rate your current daily stress level?': 'Stress'
})
```

```python
df[['Gender', 'Age', 'Usage', 'Stress']].head()
```

```python
df['Usage'] = pd.to_numeric(df['Usage'], errors='coerce')
df['Stress'] = pd.to_numeric(df['Stress'], errors='coerce')

df = df.dropna(subset=['Gender', 'Age', 'Usage', 'Stress'])

df[['Gender', 'Age', 'Usage', 'Stress']].head()
```

```python
print(df[['Gender', 'Age', 'Usage', 'Stress']].info())
print()
print(df[['Usage', 'Stress']].describe())
```

```python
plt.figure(figsize=(8,5))
plt.hist(df['Usage'], bins=15)
plt.title("Distribution of Social Media Usage")
plt.xlabel("Daily social media usage (hours)")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.hist(df['Stress'], bins=15)
plt.title("Distribution of Stress Level")
plt.xlabel("Stress level")
plt.ylabel("Frequency")
plt.show()
```

```python
usage_by_gender = df.groupby('Gender')['Usage'].mean()
usage_by_gender
```

```python
plt.figure(figsize=(8,5))
usage_by_gender.plot(kind='bar')
plt.title("Average Social Media Usage by Gender")
plt.xlabel("Gender")
plt.ylabel("Average usage (hours)")
plt.show()
```

```python
stress_by_age = df.groupby('Age')['Stress'].mean()
stress_by_age
```

```python
plt.figure(figsize=(8,5))
stress_by_age.plot(kind='bar')
plt.title("Average Stress Level by Age Group")
plt.xlabel("Age group")
plt.ylabel("Average stress")
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.scatter(df['Usage'], df['Stress'])
plt.title("Social Media Usage vs Stress Level")
plt.xlabel("Usage (hours)")
plt.ylabel("Stress level")
plt.show()
```

```python
corr, p_corr = pearsonr(df['Usage'], df['Stress'])

print("Correlation coefficient:", corr)
print("Correlation p-value:", p_corr)
```

```python
male = df[df['Gender'] == 'Male']['Stress']
female = df[df['Gender'] == 'Female']['Stress']

t_stat, p_ttest = ttest_ind(male, female, nan_policy='omit')

print("T-test statistic:", t_stat)
print("Gender t-test p-value:", p_ttest)
```

```python
groups = [group['Stress'].values for name, group in df.groupby('Age')]

anova_result = f_oneway(*groups)

print("ANOVA statistic:", anova_result.statistic)
print("Age ANOVA p-value:", anova_result.pvalue)
```

```python
if p_corr < 0.05:
    print("There is a statistically significant relationship between social media usage and stress level.")
else:
    print("There is no statistically significant relationship between social media usage and stress level.")

if p_ttest < 0.05:
    print("Stress levels differ significantly between genders.")
else:
    print("Stress levels do not differ significantly between genders.")

if anova_result.pvalue < 0.05:
    print("Stress levels differ significantly across age groups.")
else:
    print("Stress levels do not differ significantly across age groups.")
```
